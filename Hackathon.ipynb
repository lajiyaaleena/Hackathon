```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

AGE-GROUP VS LITERATE

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_new, x='Age-group', y='Literate persons', ci=None)
plt.title("Literate people in Karnataka")
plt.xlabel("Age-group")
plt.ylabel("Literate people")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Checking whether there are any age ranges where on gender is more educated than the other or consistently outperforms the other
```

```python
plt.figure(figsize=(10, 6))

#create a combined df of district code and working male and female data
combined_df = df_new.melt(id_vars=['Age-group'], value_vars=['Literate males','Literate females'],
                      var_name='Gender', value_name='Count')

# Create a bar plot with 'District code' on the x-axis and 'Count' on the y-axis, differentiated by 'Gender'
sns.barplot(x='Age-group', y='Count', hue='Gender', data=combined_df, ci=None)

# Add labels and title
plt.xlabel('Age group')
plt.ylabel('Number of Individuals')
plt.title('Literate males and females')
plt.xticks(rotation=45)
plt.legend()
# Display the legend
plt.legend().remove()

# Create a custom legend
legend_labels = ['Number of Males', 'Number of Females']
plt.legend(legend_labels, title='Gender')

# Show the plot
plt.show()
```

PLOT DISTRICT VS LITERATE PEOPLE

```python
df1=df[df['Dist code']!=0]
```

```python

district_literate = df1[['Dist code', 'Literate persons']]

# Grouping by District code and summing up literate persons
district_literate_grouped = district_literate.groupby('Dist code')['Literate persons'].sum().reset_index()

```

```python
# Sorting the data for better visualization
district_literate_grouped = district_literate_grouped.sort_values(by='Literate persons', ascending=False)

```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.bar(district_literate_grouped['Dist code'], district_literate_grouped['Literate persons'], color='green')
plt.xlabel('District code')
plt.ylabel('Literate persons')
plt.title('District code vs Literate people')
plt.xticks(rotation=90)
plt.tight_layout()

# Display the plot
plt.show()
```

DISTRICT VS EMPLOYED PEOPLE

```python
df2=df2[df2['District code']!=0]
```

```python
district_employed = df2[['District code', 'Main workers(persons)']]

# Grouping by District code and summing up literate persons
district_employed_grouped = district_employed.groupby('District code')['Main workers(persons)'].sum().reset_index()

```

```python
# Sorting the data for better visualization
district_employed_grouped = district_employed_grouped.sort_values(by='Main workers(persons)', ascending=False)

```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.bar(district_employed_grouped['District code'], district_employed_grouped['Main workers(persons)'], color='blue')
plt.xlabel('District code')
plt.ylabel('Employed persons')
plt.title('District code vs Employed people')
plt.xticks(rotation=90)
plt.tight_layout()

# Display the plot
plt.show()
```

PIE-CHART OF OVERALL LITERACY DISTRIBUTION
